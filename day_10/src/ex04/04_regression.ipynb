2. Run the function for `LinearRegression`, `DecisionTreeRegressor`, `RandomForestRegressor`. You may choose the parameters by yourself, but find the good enough for you.

````python
def crossval(n_splits, X, y, model):
    kf = KFold(n_splits=n_splits, random_state=21, shuffle=True)
    rmse_values = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        rmse = np.sqrt(mean_squared_error(y_test, predictions))
        rmse_values.append(rmse)
        print(f"train - {np.mean(rmse_values):.2f} | test - {rmse:.2f}")
    
    print(f"Average RMSE on crossval is {np.mean(rmse_values):.2f}")
````

````python
linear_reg = LinearRegression()
decision_tree = DecisionTreeRegressor(random_state=21)
random_forest = RandomForestRegressor(n_estimators= 100, random_state=21)

crossval(n_splits=5, X=X, y=y, model=linear_reg)
crossval(n_splits=5, X=X, y=y, model=decision_tree)
crossval(n_splits=5, X=X, y=y, model=random_forest)
````

## 3. Predictions and evaluation

1. Make predictions for the test dataset using each of the three models with the finalized parameters.

````python
y_pred_linear = linear_reg.predict(X_test)
y_pred_decision_tree = decision_tree.predict(X_test)
y_pred_random_forest = random_forest.predict(X_test)

actual_avg_diff = np.mean(y_test - y_test.mean())

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_linear, alpha=0.5, label='Linear Regression')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Actual Average Difference')
plt.ylabel('Predicted Value')
plt.legend()
plt.title('Linear Regression Predictions')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_decision_tree, alpha=0.5, label='Decision Tree')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Actual Average Difference')
plt.ylabel('Predicted Value')
plt.legend()
plt.title('Decision Tree Predictions')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_random_forest, alpha=0.5, label='Random Forest')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Actual Average Difference')
plt.ylabel('Predicted Value')
plt.legend()
plt.title('Random Forest Predictions')
plt.show()
````
